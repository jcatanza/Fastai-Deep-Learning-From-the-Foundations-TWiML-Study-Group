## 2. Refine the optimizer

In **PyTorch**, the base optimizer in `torch.optim` is just a dictionary that stores the hyper-parameters and references to the parameters of the model we want to train in **parameter groups** (different groups can have different learning rates/momentum/weight decay... this is what lets us implement **discriminative learning rates**).

It contains a method `step` that updates parameters via the gradients and a method `zero_grad` to **detach** and **zero** the gradients of all our parameters.

We build the equivalent from scratch, only ours will be more flexible. In our implementation, the step function loops over all the parameters to execute `step` using `stepper` functions that we have to provide when initializing the optimizer.

[Jump_to lesson 11 video](https://course.fast.ai/videos/?lesson=11&t=4074)

````python
class Optimizer():
    def __init__(self, params, steppers, **defaults):
        
        # params might be a generator
        self.param_groups = list(params)
        
        # ensure param_groups is a list of lists
        if not isinstance(self.param_groups[0], list): 
            self.param_groups = [self.param_groups]
            
        # hypers is a list containing a dictionary (which is read from defaults) for each param_group
        self.hypers = [{**defaults} for p in self.param_groups]
        
        # make steppers a list of objects
        self.steppers = listify(steppers)

    # method that returns a list containing a tuple of 
    #      (param_group, and its corresponding hyperparameter dictionary) 
    #      for each param_group
    def grad_params(self):
        return [(p,hyper) for pg,hyper in zip(self.param_groups,self.hypers)
            for p in pg if p.grad is not None]

    # method to initialize each parameter to zero, after detaching it from the computation graph
    def zero_grad(self):
        for p,hyper in self.grad_params():
            p.grad.detach_()
            p.grad.zero_()

    # for each parameter, compute the next step from the sequential composition of a list of 
    #      methods provided in steppers.
    def step(self):
        for p,hyper in self.grad_params(): compose(p, self.steppers, **hyper)
````

### Compute a basic SGD step

````python
#export
# compute the SGD step
def sgd_step(p, lr, **kwargs):
    p.data.add_(-lr, p.grad.data)
    return p
````

````python
# create an SGD optimizer function, by using the Optimizer class
#     with the `steppers` list containing only `sgd_step`
opt_func = partial(Optimizer, steppers=[sgd_step])
````

### refactor the Recorder, ParamScheduler, and LR_Find callbacks
Now that we have changed the optimizer, we will need to adjust the callbacks that were using properties from the PyTorch optimizer: in particular the hyper-parameters are in the list of dictionaries `opt.hypers` (PyTorch has everything in the the list of param groups).

````python
#export

# create Recorder callback 
class Recorder(Callback):
    def begin_fit(self): self.lrs,self.losses = [],[]

    def after_batch(self):
        if not self.in_train: return
        # append the learning rate to the learning rate hyperparameter list
        self.lrs.append(self.opt.hypers[-1]['lr'])
        # detach the loss function from the computation graph and put it on CPU and append it to a list
        self.losses.append(self.loss.detach().cpu())        

    # methods to plot learning rate and loss as a function of training step
    def plot_lr  (self): plt.plot(self.lrs)
    def plot_loss(self): plt.plot(self.losses)
        
    # method to plot learning rate against loss during training
    def plot(self, skip_last=0):
        losses = [o.item() for o in self.losses]
        n = len(losses)-skip_last
        plt.xscale('log')
        plt.plot(self.lrs[:n], losses[:n])

# create ParamScheduler callback
class ParamScheduler(Callback):
    _order=1
    
    # intitialize parameter name and list of scheduler functions
    def __init__(self, pname, sched_funcs):
        self.pname,self.sched_funcs = pname,listify(sched_funcs)

    def begin_batch(self): 
        
        # check that we are in training
        if not self.in_train: return
        
        # fractional number of current epoch in the training process
        pos = self.n_epochs/self.epochs
        
        # create the schedule function
        fs = self.sched_funcs
        if len(fs)==1: 
            fs = fs*len(self.opt.param_groups)
            
        # add a scheduler function for this parameter to the hyperparamter list
        for f,h in zip(fs,self.opt.hypers): 
            h[self.pname] = f(pos)

# create LR_find callback
class LR_Find(Callback):
    _order=1
    
    # initialize hyperparameters
    def __init__(self, max_iter=100, min_lr=1e-6, max_lr=10):
        self.max_iter,self.min_lr,self.max_lr = max_iter,min_lr,max_lr
        self.best_loss = 1e9
        
    def begin_batch(self): 
        
        # check that we are in training
        if not self.in_train: return
        
        # fractional number of current batch in the training process
        pos = self.n_iter/self.max_iter
        
        # update the learning rate according to the current schedule
        #     and store it in the hyperparameter dictionary
        lr = self.min_lr * (self.max_lr/self.min_lr) ** pos
        for pg in self.opt.hypers: 
            pg['lr'] = lr
            
    def after_step(self):
        
        # raise an exception that cancels training if either the current batch number or the loss are outside their valid ranges
        if self.n_iter>=self.max_iter or self.loss>self.best_loss*10:
            raise CancelTrainException()
        
        # update the best loss
        if self.loss < self.best_loss: 
            self.best_loss = self.loss
````

## 3. Weight Decay

[Jump_to lesson 11 video](https://course.fast.ai/videos/?lesson=11&t=4623)

By letting our model learn high parameters, it might fit all the data points in the training set with an over-complex function that has very sharp changes, which will lead to overfitting.

<img src="images/overfit.png" alt="Fitting vs over-fitting" width="600">

Weight decay comes from the idea of L2 regularization, which consists in adding to your loss function the sum of all the weights squared. Why do that? Because when we compute the gradients, it will add a contribution to them that will encourage the weights to be as small as possible.

Limiting our weights from growing too much is going to hinder the training of the model, but it will improve model generalizability. Going back to the theory a little bit, weight decay (or just `wd`) is a parameter that controls that sum of squares we add to our loss:
``` python
loss_with_wd = loss + (wd/2) * (weights**2).sum()
```

In practice though, it would be very inefficient (and maybe numerically unstable) to compute that big sum and add it to the loss. 

In the grand scheme things, it is the **derivatives of the loss with respect to the weights** (rather than the loss itself) that figures into the parameter updates.  If you remember a little bit of high school math, the derivative of `p**2` with respect to `p` is `2*p`. So adding that big sum to the loss is effectively the same -- when it comes to computing the parameter updates -- as adding $wd \mult weight$ to the gradient of the loss:
``` python
weight.grad += wd * weight
```

for every weight in our model, which in the case of vanilla SGD is equivalent to updating the parameters with:
``` python
updated_weight = weight - lr*(weight.grad + wd*weight)
```

This technique of modifying the parameter updates instead of the loss functiono is called "weight decay", as each weight is decayed by a factor `lr * wd`, as it's shown in this last formula.

This only works for standard SGD, as we have seen that with momentum, RMSProp and Adam, the update has some additional formulas around the gradient. In those cases, the formula that comes from L2 regularization:
``` python
weight.grad += wd * weight
```
is **different** than weight decay
``` python
updated_weight = weight - lr * weight.grad - lr * wd * weight
```
*Note: the above statement is unclear. It would be great to see an example.* 

Most libraries use the first one, but as it was pointed out in [Decoupled Weight Regularization](https://arxiv.org/pdf/1711.05101.pdf) by Ilya Loshchilov and Frank Hutter, it is better to use the second one with the Adam optimizer, which is why fastai made it its default.

Weight decay is subtracting `lr*wd*weight` from the weights. We need this function to have an attribute `_defaults` so that we are sure there is an hyper-parameter of the same name in our `Optimizer`.

````python
#export
def weight_decay(p, lr, wd, **kwargs):
    p.data.mul_(1 - lr*wd)
    return p
weight_decay._defaults = dict(wd=0.)
````

L2 regularization is adding `wd*weight` to the gradients.

````python
#export
def l2_reg(p, lr, wd, **kwargs):
    p.grad.data.add_(wd, p.data)
    return p
l2_reg._defaults = dict(wd=0.)
````

Let's allow steppers to add to our `defaults` (which are the default values of all the hyper-parameters). This helper function adds in `dest` the key/values it finds while going through `os` and applying `f` when they was no `key` of the same name.

````python
#export
# This function needs clear documentation of the inputs and outputs
#       so the user doesn't have to **guess** at its syntax
# Here are my guesses
# `dest` is a dict()
# `f` is a function that produces a dictionary of hyperparameters from an input `o`
#      (which presumably specifies a parameter group)
# `os` is a list of `steppers` that can be inputs to the function `f`, therefore parameter groups?
# `maybe_update()` checks whether `os` has any hyperparameters not already in `dest`, and if so, addes them 
def maybe_update(os, dest, f):
    for o in os:
        for k,v in f(o).items():
            if k not in dest: dest[k] = v

# getattr method to get d._defaults from the input object d, returns an empty set if not found 
def get_defaults(d): 
    return getattr(d,'_defaults',{})
````

This is the same as before, we just take the default values of the steppers when none are provided in the kwargs.

````python
#export
class Optimizer():
    def __init__(self, params, steppers, **defaults):
        
        self.steppers = listify(steppers)
        
        # this is the only difference from previous version
        maybe_update(self.steppers, defaults, get_defaults)
        
        # might be a generator
        self.param_groups = list(params)
        
        # ensure params is a list of lists
        if not isinstance(self.param_groups[0], list): 
            self.param_groups = [self.param_groups]
            
        self.hypers = [{**defaults} for p in self.param_groups]

    def grad_params(self):
        return [(p,hyper) for pg,hyper in zip(self.param_groups,self.hypers)
            for p in pg if p.grad is not None]

    def zero_grad(self):
        for p,hyper in self.grad_params():
            p.grad.detach_()
            p.grad.zero_()

    def step(self):
        for p,hyper in self.grad_params(): compose(p, self.steppers, **hyper)
            
````

````python
#export
# create a SGD optimizer with weight decay 
sgd_opt = partial(Optimizer, steppers=[weight_decay, sgd_step])
````

## 4. Momentum

[Jump_to lesson 11 video](https://course.fast.ai/videos/?lesson=11&t=4872)

**Momentum is a general term for techniques that factor prior behavior of the gradients into determining the next step. Plain SGC (Stochastic Gradient Descent) computes the next step using only the loss function gradients from the current batch. Momentum techniques additionally use 'moving averages' of gradients, which take into account the gradient values at previous batches. Momentum comes in several flavors, some of which we will investigate in the sections below.**

Momentum requires us to add 'state' variables to the optimizer in order to keep track the moving averages. We need to save the moving average of the gradients to be able to compute the step and store this inside the optimizer state. To do this, we introduce the `Stat` object, which has two methods:
- `init_state`, that returns the initial state (a tensor of 0. for the moving average of gradients)
- `update`, that updates the state with the new gradient value

We also need the `_defaults` values of those objects, to allow them to provide default values to hyper-parameters.

````python
#export
class StatefulOptimizer(Optimizer):
    def __init__(self, params, steppers, stats=None, **defaults): 
        self.stats = listify(stats)
        maybe_update(self.stats, defaults, get_defaults)
        super().__init__(params, steppers, **defaults)
        self.state = {}
        
    def step(self):
        for p,hyper in self.grad_params():
            if p not in self.state:
                #Create a state for p and call all the statistics to initialize it.
                self.state[p] = {}
                maybe_update(self.stats, self.state[p], lambda o: o.init_state(p))
            state = self.state[p]
            for stat in self.stats: state = stat.update(p, state, **hyper)
            compose(p, self.steppers, **state, **hyper)
            self.state[p] = state
````

````python
#export
class Stat():
    _defaults = {}
    def init_state(self, p): 
        raise NotImplementedError
    def update(self, p, state, **kwargs): 
        raise NotImplementedError    
````

Here is an example of a class that inherits from `Stat`:

````python
class AverageGrad(Stat):
        
    _defaults = dict(mom=0.9)

    # initialize a dictionary with key 'grad_avg' and value of zeros
    def init_state(self, p):
        # Q: why isn't the dictionary called 'state'
        return {'grad_avg': torch.zeros_like(p.grad.data)}
    
    def update(self, p, state, mom, **kwargs):
        
        # multiply state['grad_avg'] in place by mom and add in place p.grad.data
        # Q: where are state and state['grad_avg'] defined
        state['grad_avg'].mul_(mom).add_(p.grad.data)
        return state
````

### Regular momentum

Here is a method to compute the momentum step.

````python
#export
def momentum_step(p, lr, grad_avg, **kwargs):
    # compute the additive step, -lr*grad_avg, for the parameter update 
    # same as SGD step except that update is computed from the current EWMA of the gradient 
    #      instead of from the gradient for the current batch 
    p.data.add_(-lr, grad_avg)
    return p
````

## 5. Momentum experiments

What does momentum do to the gradients exactly? Let's do some plots to find out!

[Jump_to lesson 11 video](https://course.fast.ai/videos/?lesson=11&t=5487)

````python
# sample 200 points from a Gaussian with mean 0.3 and unit variance
#      put the points on a uniform grid of 200 points on the interval [-4., 4.]
x = torch.linspace(-4, 4, 200)
y = torch.randn(200) + 0.3
# grid of momentum values
betas = [0.5, 0.7, 0.9, 0.99]
````

````python
# produce a family of moving average plots, with each plot corresponding to a given value of momentum 
def plot_mom(f):
    _,axs = plt.subplots(2,2, figsize=(12,8))
    # family of subplots, one for each momentum value
    for beta,ax in zip(betas, axs.flatten()):
        ax.plot(y, linestyle='None', marker='.')
        # initialize moving average to None and the list of updated moving averages `avgs` to []
        avg,avgs = None,[]
        for i,yi in enumerate(y):
            # given the current value of the moving average `avg`,
            #      f(avg, beta, yi,i) computes and returns the tuple avg, avg2:
            #           avg is the the updated moving average of the `yi` with momentum `beta`
            #           avg2 is either a second copy of the updated moving average when f() is mom1() or mom2(), 
            #                or a debiased updated moving average when f() is mom3()
            avg, avg2 = f(avg, beta, yi, i)
            # appends the updated or updated and debiased `avg` to the cumulative list of moving averages
            #      passes the updated (but not debiased) `avg` to the next call of f()
            avgs.append(avg2)
        ax.plot(avgs, color='red')
        ax.set_title(f'beta={beta}')
        
````

### Regular momentum

````python
# momentum
def mom1(avg, beta, yi, i):
    # initialize avg to y0 at the first data point
    # return two copies of the updated average
    if avg is None: 
        avg=yi
    else:
        #update the moving average
        avg = beta*avg + yi
    return avg, avg
plot_mom(mom1)
````

As we can see, with a too high value, it may go way too high with no way to change its course.

### Exponentially weighted moving average (EWMA)
Another way to smooth noisy data is to form an exponentially weighted moving average of the data. The next value in the moving average is $\beta$ times the current average plus $(1 - \beta)$ times the value of the current data point. In this case, there is a dampening of $(1-\beta)$ in front of the current data value, which is "less trusted" than the current average if $0.5 < \beta < 1$. We'll define `lin_comb` (*linear combination*) to make this easier (note that in the lesson this was named `ewma`).

````python
#export
# linear interpolation
# this is an affine transformation if 0 <= beta <= 1
def lin_comb(v1, v2, beta): return beta*v1 + (1-beta)*v2
````

````python
def mom2(avg, beta, yi, i):
    # initialize avg to y0 at the first data point
    # return two copies of the updated average
    if avg is None: avg=yi
    else:
        #update the moving average
        avg = lin_comb(avg, yi, beta)
    return avg, avg
plot_mom(mom2)
````

We can see it gets to a zero-constant when the data is purely random. If the data has a certain shape, it will get that shape (with some delay for high beta).

````python
y = 1 - (x/3) ** 2 + torch.randn(200) * 0.1
````

````python
y[0]=0.5
````

EWMA begins to lag the data for large values of $\beta$ and as beta becomes larger than 0.9, the EWMA becomes attenuated in amplitude as well.

````python
plot_mom(mom2)
````

### Debiased EWMA

Debiasing is here to correct the wrong information we may have in the very first batch. The debias term corresponds to the sum of the coefficient in our moving average. At the time step i, our average is:

$\begin{align*}
avg_{i} &= \beta\ avg_{i-1} + (1-\beta)\ v_{i} = \beta\ (\beta\ avg_{i-2} + (1-\beta)\ v_{i-1}) + (1-\beta)\ v_{i} \\
&= \beta^{2}\ avg_{i-2} + (1-\beta)\ \beta\ v_{i-1} + (1-\beta)\ v_{i} \\
&= \beta^{3}\ avg_{i-3} + (1-\beta)\ \beta^{2}\ v_{i-2} + (1-\beta)\ \beta\ v_{i-1} + (1-\beta)\ v_{i} \\
&\vdots \\
&= (1-\beta)\ \beta^{i}\ v_{0} + (1-\beta)\ \beta^{i-1}\ v_{1} + \cdots + (1-\beta)\ \beta^{2}\ v_{i-2} + (1-\beta)\ \beta\  v_{i-1} + (1-\beta)\ v_{i}
\end{align*}$

and so the sum of the coefficients is

$\begin{align*}
S &=(1-\beta)\ \beta^{i} + (1-\beta)\ \beta^{i-1} + \cdots + (1-\beta)\ \beta^{2} + (1-\beta)\ \beta + (1-\beta) \\
&= (\beta^{i} - \beta^{i+1}) + (\beta^{i-1} - \beta^{i}) + \cdots + (\beta^{2} - \beta^{3}) + (\beta - \beta^{2}) + (1-\beta) \\
&= 1 - \beta^{i+1}
\end{align*}$

since all the other terms cancel out each other.

By dividing by this term (debiasing), we make our moving average a **true weighted average** of all the data points, in the sense that all the weights sum up to 1.

````python
# add debiasing
def mom3(avg, beta, yi, i):
    if avg is None: 
        # initializing avg to 0 doesn't make sense, but it works better than yi
        avg=0
    else:
        avg = lin_comb( avg, yi, beta )
    # at each step, debiasing the avg makes it a weighted average of the data points 
    #      where the weights sum to one. In other words, an affine transformation.
    return avg, avg/(1-beta**(i+1))
plot_mom(mom3)
````

The debiased EWMA still lags the data as $\beta$ increases but is less attenuated in amplitude for large $\beta$

## 6. Adam and friends

In Adam, we use the gradient averages but with dampening (not like in SGD with momentum), so let's add this to the `AverageGrad` class.

[Jump_to lesson 11 video](https://course.fast.ai/videos/?lesson=11&t=5889)

````python
#export
class AverageGrad(Stat):
    _defaults = dict(mom=0.9)
    
    def __init__(self, dampening:bool=False): self.dampening=dampening
    def init_state(self, p): return {'grad_avg': torch.zeros_like(p.grad.data)}
    def update(self, p, state, mom, **kwargs):
        state['mom_damp'] = 1-mom if self.dampening else 1.
        state['grad_avg'].mul_(mom).add_(state['mom_damp'], p.grad.data)
        return state
````

We also need to track the moving average of the gradients squared.

````python
#export
class AverageSqrGrad(Stat):
    _defaults = dict(sqr_mom=0.99)
    
    def __init__(self, dampening:bool=True): self.dampening=dampening
    def init_state(self, p): return {'sqr_avg': torch.zeros_like(p.grad.data)}
    def update(self, p, state, sqr_mom, **kwargs):
        state['sqr_damp'] = 1-sqr_mom if self.dampening else 1.
        state['sqr_avg'].mul_(sqr_mom).addcmul_(state['sqr_damp'], p.grad.data, p.grad.data)
        return state
````

Note on `addcmul_()` syntax:

`addcmul(tensor,value,tensor1,tensor2) = tensor + valuetensor1tensor2`

If unspecified,`value` is taken to be 1 by default

`addcmul_()` performs the operation in place

We will also need the number of steps done during training for the debiasing.

````python
#export
class StepCount(Stat):
    
    # initialize state
    # Q: shouldnt we initialize by setting state=dict() and state['step'] = 0?
    def init_state(self, p): return {'step': 0}
    
    # update state
    def update(self, p, state, **kwargs):
        state['step'] += 1
        return state
````

This helper function computes the debias term. If we include dampening (`damp = 1 - mom`), we get the same result as before. 

If we don't use dampening (`damp = 1`), we will need to divide by `1 - mom` because that term is missing everywhere.

Q: this seems to be implemented **backwards**, i.e. semantically no dampening should give the previous result for debiasing, while including dampening should divide the previous result by `1-mom`

Momentum dampening seems to be multiplying the debiased step by a factor of `1-m`. The effect is to slow down (or 'dampen') the parameter updates.

````python
#export
def debias(mom, damp, step): return damp * (1 - mom**step) / (1-mom)
````

Then the Adam step is just the following:

````python
#export
# eps controls the strength of the feature
# Jeremy says eps should be between 0.1 and 0.001, but most people use 1e-7, which Jeremy says makes no sense
def adam_step(p, lr, mom, mom_damp, step, sqr_mom, sqr_damp, grad_avg, sqr_avg, eps, **kwargs):
    debias1 = debias(mom,     mom_damp, step)
    debias2 = debias(sqr_mom, sqr_damp, step)
    # debiased, dampened momentum
    p.data.addcdiv_(-lr / debias1, grad_avg, (sqr_avg/debias2).sqrt() + eps)
    return p
adam_step._defaults = dict(eps=1e-5)
````

Note on `addcdiv_()` syntax:

`addcdiv(tensor,value,tensor1,tensor2) = tensor + value(tensor1/tensor2)`

If unspecified,`value` is taken to be 1 by default 

`addcdiv_()` performs the operation in place

````python
#export
def adam_opt(xtra_step=None, **kwargs):
    return partial(StatefulOptimizer, steppers=[adam_step,weight_decay]+listify(xtra_step),
                   stats=[AverageGrad(dampening=True), AverageSqrGrad(), StepCount()], **kwargs)
````

````python
learn,run = get_learn_run(nfs, data, 0.001, conv_layer, cbs=cbfs, opt_func=adam_opt())
````

### Adam improves the model fit tremendously over the baseline

````python
run.fit(3, learn)
````

## 7. LAMB 
### essentially Adam, with all steps averaged over a layer

[Jump_to lesson 11 video](https://course.fast.ai/videos/?lesson=11&t=6038)

It's then super easy to implement a new optimizer. This is LAMB from a [very recent paper](https://arxiv.org/pdf/1904.00962.pdf):

with slightly refactored math:

$\begin{align}
g_{t}^{l} &= \nabla L(w_{t-1}^{l}, x_{t}) \\
m_{t}^{l} &= \beta_{1} m_{t-1}^{l} + (1-\beta_{1}) g_{t}^{l} \\
v_{t}^{l} &= \beta_{2} v_{t-1}^{l} + (1-\beta_{2}) g_{t}^{l} \odot g_{t}^{l} \\
m_{t}^{l} &= m_{t}^{l} / (1 - \beta_{1}^{t}) \\
v_{t}^{l} &= v_{t}^{l} / (1 - \beta_{2}^{t}) \\
r_{1} &= \|w_{t-1}^{l}\|_{2} \\
s_{t}^{l} &= \frac{m_{t}^{l}}{\sqrt{v_{t}^{l} + \epsilon}} + \lambda w_{t-1}^{l} \\ 
r_{2} &= \| s_{t}^{l} \|_{2} \\
\eta^{l} &= \eta * r_{1}/r_{2} \\ 
w_{t}^{l} &= w_{t}^{l-1} - \eta_{l} * s_{t}^{l} \\
\end{align}$

````python
def lamb_step(p, lr, mom, mom_damp, step, sqr_mom, sqr_damp, grad_avg, sqr_avg, eps, wd, **kwargs):
    # compute debias factors for mom and sqr_mom
    debias1 = debias(mom,     mom_damp, step)
    debias2 = debias(sqr_mom, sqr_damp, step)
    # compute rms value of p.data
    r1 = p.data.pow(2).mean().sqrt()
    # compute adam step plus weight decay
    step = (grad_avg/debias1) / ((sqr_avg/debias2).sqrt()+eps) + wd*p.data
    # compute rms step (presumably averaged over the channels in the layer)
    r2 = step.pow(2).mean().sqrt()
    # add to p.data the step -l*multiplier, 
    #    where multiplier is the minimum of r1/r2 and 10
    #   r1/r2 is the ratio of the rms data and the rms step
    p.data.add_(-lr * min(r1/r2,10), step)
    return p
lamb_step._defaults = dict(eps=1e-6, wd=0.)
````

````python
lamb = partial(StatefulOptimizer, steppers=lamb_step, stats=[AverageGrad(dampening=True), AverageSqrGrad(), StepCount()])
````

````python
learn,run = get_learn_run(nfs, data, 0.003, conv_layer, cbs=cbfs, opt_func=lamb)
````

### LAMB is about the same as the baseline

````python
run.fit(3, learn)
````
